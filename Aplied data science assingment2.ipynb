```python
import pandas as pd

def read_worldbank_data(filename):
    # Read the CSV file into a pandas DataFrame
    data = pd.read_csv(filename)

    # Transpose the DataFrame to have years as columns and countries as rows
    data_transposed = data.melt(id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='Value')

    # Pivot the transposed DataFrame to get countries as columns
    countries_as_columns = data_transposed.pivot_table(index=['Series Name', 'Series Code', 'Year'], columns='Country Name', values='Value').reset_index()

    # Pivot the transposed DataFrame to get years as columns
    years_as_columns = data_transposed.pivot_table(index=['Series Name', 'Series Code', 'Country Name'], columns='Year', values='Value').reset_index()

    return countries_as_columns, years_as_columns

# Example usage:
filename = 'data.csv'
countries_df, years_df = read_worldbank_data(filename)

# Clean the transposed DataFrame
countries_df = countries_df.dropna(axis=1, how='all')  # Remove columns with all NaN values
years_df = years_df.dropna(axis=1, how='all')  # Remove columns with all NaN values

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Load your dataset
data = pd.read_csv('data.csv')  # Replace 'data.csv' with your file name

# Extract unique indicators
unique_indicators = data['Series Name'].unique()

# Randomly select 10 countries
all_countries = data['Country Name'].unique()
selected_countries = random.sample(list(all_countries), 10)

# Filter data for unique indicators and selected countries
selected_data = data[data['Series Name'].isin(unique_indicators) & data['Country Name'].isin(selected_countries)]

# Extract columns related to years
year_columns = [col for col in data.columns if col.startswith('20')]

# Melt the data to restructure years into a single column
selected_data_melted = selected_data.melt(id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], value_vars=year_columns, var_name='Year', value_name='Value')

# Convert 'Year' column to numeric for sorting
selected_data_melted['Year'] = selected_data_melted['Year'].str.extract('(\d+)', expand=False).astype(int)

# Visualize Time Series Trends for Indicators
for indicator in unique_indicators:
    plt.figure(figsize=(10, 6))
    for country in selected_countries:
        country_data = selected_data_melted[(selected_data_melted['Country Name'] == country) & (selected_data_melted['Series Name'] == indicator)]
        plt.plot(country_data['Year'], country_data['Value'], label=country)
    plt.xlabel('Year')
    plt.ylabel(indicator)
    plt.title(f'{indicator} Trends')
    plt.legend()
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

# Load your dataset
data = pd.read_csv('data.csv')  # Replace 'data.csv' with your file name

# Extract two unique series names
unique_series_names = data['Series Name'].unique()[:2]

# Randomly select two countries
all_countries = data['Country Name'].unique()
selected_countries = random.sample(list(all_countries), 2)

# Filter data for the selected series names and countries
selected_data = data[data['Series Name'].isin(unique_series_names) & data['Country Name'].isin(selected_countries)]

# Extract columns related to years
year_columns = [col for col in data.columns if col.startswith('20')]

# Melt the data to restructure years into a single column
selected_data_melted = selected_data.melt(id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], value_vars=year_columns, var_name='Year', value_name='Value')

# Convert 'Year' column to numeric for sorting
selected_data_melted['Year'] = selected_data_melted['Year'].str.extract('(\d+)', expand=False).astype(int)

# Prepare data for plotting
plt_data = selected_data_melted.pivot_table(index=['Country Name', 'Year'], columns='Series Name', values='Value').reset_index()

# Plotting the trends for the two selected series in the two countries using a bar plot
plt.figure(figsize=(10, 6))

for series_name in unique_series_names:
    for country in selected_countries:
        country_data = plt_data[(plt_data['Country Name'] == country)]
        plt.bar(country_data[country_data['Country Name'] == country]['Year'], country_data[country_data['Country Name'] == country][series_name], label=f'{series_name} - {country}', alpha=0.7)

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Comparison of Two Series in Two Countries')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv('data.csv')  # Replace 'data.csv' with your file name

# Extract three unique series names (indicators)
unique_series_names = data['Series Name'].unique()[:3]

# Prepare a DataFrame for each indicator to find top 10 countries
top_countries = {}
for indicator in unique_series_names:
    # Filter data for the current indicator and calculate the mean value over all years for each country
    indicator_data = data[data['Series Name'] == indicator]
    indicator_data['Mean'] = indicator_data.iloc[:, 4:].mean(axis=1)  # Calculate mean excluding non-year columns

    # Sort by mean value in descending order
    sorted_indicator_data = indicator_data.sort_values(by='Mean', ascending=False)

    # Select top 10 countries with the highest mean value
    top_countries[indicator] = sorted_indicator_data['Country Name'].head(10).tolist()

# Plotting the data for the top 10 countries having the highest overall values for each indicator
plt.figure(figsize=(15, 10))

for i, indicator in enumerate(unique_series_names, 1):
    plt.subplot(3, 1, i)
    for country in top_countries[indicator]:
        country_data = data[(data['Country Name'] == country) & (data['Series Name'] == indicator)]
        plt.plot(country_data.columns[4:], country_data.iloc[0, 4:], label=f'{country}')

    plt.xlabel('Year')
    plt.ylabel('Value')
    plt.title(f'Top 10 Countries with Highest Overall Values - {indicator}')
    plt.legend()
    plt.tight_layout()

plt.show()
```
